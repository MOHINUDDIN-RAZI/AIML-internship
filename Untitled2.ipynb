```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt

# Load the data from the Excel file
data = pd.read_excel("insurance.xlsx")
df1 = pd.DataFrame(data)

# Drop the specified columns
df1 = df1.drop(["sex", "smoker", "region"], axis=1)

# Drop rows with NaN values
df1 = df1.dropna()

# Drop duplicate rows
df1 = df1.drop_duplicates()

# Reset the index
df1.reset_index(inplace=True, drop=True)

# Create a scatter plot of 'age' vs. 'charges'
plt.scatter(df1['age'], df1['charges'], color='red')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.title('Scatter Plot of Age vs. Charges')
plt.show()

# Linear Regression
regr = linear_model.LinearRegression()
x = df1[['age']]
y = df1['charges']

regr.fit(x, y)

print('Intercept: \n', regr.intercept_)
print('Coefficient: \n', regr.coef_[0])

# Predict charges for a given age (e.g., 30)
age_to_predict = 30
predicted_charges = regr.predict([[age_to_predict]])
print(f"Predicted Charges for Age {age_to_predict}: {predicted_charges[0]}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import export_text

# Load the data from the Excel file
data = pd.read_excel("insurance.xlsx")

# Assuming you have a 'age' and 'charges' columns in your dataset
X = data[['age']].values
y = data['charges'].values

# Create and fit the Decision Tree Regressor
regressor = DecisionTreeRegressor(random_state=0)
clf = regressor.fit(X, y)

# Visualize the decision tree using text
tree_rules = export_text(clf, feature_names=['age'])
print(tree_rules)

# Predict charges for a given age (e.g., 30)
age_to_predict = np.array([[30]])
predicted_charges = regressor.predict(age_to_predict)
print(f"Predicted Charges for Age {age_to_predict[0][0]}: {predicted_charges[0]}")

# Generate a range of ages for plotting the regression line
X_grid = np.arange(min(X), max(X), 0.01)[:, np.newaxis]

# Predict the corresponding charges for the X_grid
y_grid = regressor.predict(X_grid)

# Plot the regression line
plt.scatter(X, y, color='red', label='Actual Data')
plt.plot(X_grid, y_grid, color='blue', label='Regression Line')
plt.title('Charges vs Age (Decision Tree Regression)')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.legend()
plt.savefig("decision.png")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import export_graphviz

# Load the data from the Excel file
data = pd.read_excel("insurance.xlsx")

# Assuming you have a 'Production Cost' and 'Profit' column in your dataset
X = data[['Production Cost']].values
y = data['Profit'].values

# Create and fit the Decision Tree Regressor
regressor = DecisionTreeRegressor(random_state=0)
clf = regressor.fit(X, y)

# Visualize the decision tree
plt.figure(figsize=(12, 6))
tree.plot_tree(clf)
plt.show()

# Export the decision tree to a DOT file (for visualization purposes)
export_graphviz(regressor, out_file='tree.dot', feature_names=['Production Cost'])

# Predict a value
production_cost_to_predict = 6500
predicted_profit = regressor.predict([[production_cost_to_predict]])
print("Predicted profit: %.2f" % predicted_profit)

# Generate a range of values for plotting the regression line
X_grid = np.arange(min(X), max(X), 1)[:, np.newaxis]

# Predict the corresponding profits for the X_grid
y_grid = regressor.predict(X_grid)

# Plot the regression line
plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='red', label='Actual Data')
plt.plot(X_grid, y_grid, color='blue', label='Regression Line')
plt.title('Profit vs Production Cost (Decision Tree Regression)')
plt.xlabel('Production Cost')
plt.ylabel('Profit')
plt.legend()
plt.savefig("decision.png")
plt.show()
```
